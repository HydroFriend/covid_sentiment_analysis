```python
import bz2
def read_json_file(file_path):
    with bz2.open(file_path, "rt") as r:
        response = r.read()
        response = response.replace('\n', '')
        response = response.replace('}{', '},{')
        response = "[" + response + "]"
        data = json.loads(response)
        df = pd.DataFrame(data)
        return df
```

```python


def process1D(data_path,save_path,date):
    

    hours = os.listdir(data_path)
    print(hours)
    combinelist=[]
    for i in hours:
        path_hour = os.path.join(data_path, i )
        path_min= os.listdir(path_hour)

        k = 1
        for j in path_min:
            path_final = os.path.join(path_hour, j)
            df = read_json_file(path_final)
            if k == 1:
                df_tweet = df.copy(deep=True)
                k = 0
            else:
                df_tweet = pd.concat([df_tweet, df])

        df_tweet_final = df_tweet.loc[df_tweet['text'].str.contains((words_list_total), case = False, na = False)]
        df_tweet_final = df_tweet_final.reset_index(drop=True)
        combinelist.append(df_tweet_final)
        print('process:',i)
    path = save_path + date+".csv"
    df_tweet_final=pd.concat(combinelist)
    df_tweet_final.to_csv(path, index=False)
    print('finish')
```
